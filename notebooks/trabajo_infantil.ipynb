### Trabajo Infantil

Los datos se obtuvieron de la página https://www.inegi.org.mx/programas/enti/2019/#Tabulados el día 20 de Abril del 2022.

```python
#Importando librerias necesarias
import pandas as pd
import pathlib

#Todo lo relacionado a graficar
import plotly.express as px
```

```python
#Definiendo el Path del archivo
path = pathlib.Path('..', 'data/raw/ENTI2019_Tabulados_Estatales.xlsx')
```

```python
#Leyendo los datos
data_trabajo =  pd.read_excel('../data/raw/personas_desaparecidas/enti_2019_tabulados_bas_xlsx/ENTI2019_Tabulados_Estatales.xlsx',
                                sheet_name='Cuadro 4.8', header=6)
```

```python
data_trabajo
```

```python
data_trabajo.columns
```

```python
data_trabajo = data_trabajo.rename({'Población de \n5 a 17 años\nen trabajo infantil': 'poblacion_total_trabajando',
                        'Sexo': 'hombres',
                        'Unnamed: 4': 'mujeres'},axis= 1)
data_trabajo.head()
```

```python
data_trabajo.drop('Unnamed: 2', axis=1,inplace=True)
```

```python
#Podemos remover las primeras dos filas 
data_trabajo.drop([0,1],axis=0,inplace=True)
```

```python
trabajo_por_sexo = data_trabajo[data_trabajo['Entidad federativa'] == 'Sonora'].copy()
trabajo_por_sexo.to_csv('trabajo_infantil_por_sexo')
```

```python
data_trabajo_2 =  pd.read_excel('../data/raw/personas_desaparecidas/enti_2019_tabulados_bas_xlsx/ENTI2019_Tabulados_Estatales.xlsx',
                                sheet_name='Cuadro 4.13', header=6)
```

```python
data_trabajo_2.columns
```

```python
data_trabajo_2 = data_trabajo_2.rename({'Población de \n5 a 17 años\nen ocupación\nno permitida ': 'poblacion_total_trabajando',
                        'Sector de actividad económica': 'sector_primario',
                        'Unnamed: 4': 'sector_secundario',
                        'Unnamed: 5': 'sector_terciario',
                        'Unnamed: 6':'no_especificado'},axis= 1)
data_trabajo_2.head()
```

```python
data_trabajo_2.drop('Unnamed: 2', axis=1,inplace=True)
```

```python
data_trabajo_2.drop([0,1],axis=0,inplace=True)
```

```python
trabajo_infantil_por_sector = data_trabajo_2[data_trabajo_2['Entidad federativa'] == 'Sonora'].copy()
trabajo_infantil_por_sector.to_csv('trabajo_infantil_por_sector')
```
